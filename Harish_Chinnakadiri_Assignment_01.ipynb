# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
import math
print ('r', '  ' , 'V')
for r in range(1,20,2):
  V  = 2/3 * math.pi * r**3
  print(f"{r:2}  {V:.2f}")
```

# Question 2 (20 points)

(a) Write a Python program to generate a sequence of 12 random numbers falling within the range of 50 to 150.

(b) Arrange the generated numbers in ascending order. Display the sorted list.

(c) Extend the program to arrange the numbers in descending order as well. Present the sorted list accordingly.

(d) Compute and output the average of the entire group of generated numbers.

(e) Enhance the program to calculate and display the average of the first 6 numbers from both the ascending and descending sorted lists.

```python
#a)
# Import the random module to use its functionalities
import random

# Initialize an empty list to store the random numbers
random_numbers = []

# Loop 12 times to generate 12 random numbers
for _ in range(12):
    # Generate a random number between 50 and 150 (inclusive)
    number = random.randint(50, 150)

    # Add the generated number to the list
    random_numbers.append(number)

# Print the list of random numbers
print(random_numbers)

```

```python
#b
sorted_numbers = sorted(random_numbers)
print(sorted_numbers)
```

```python
#c descending order
sorted_numbers_descending = sorted(random_numbers, reverse = True)
print(sorted_numbers_descending)
```

```python
#d average of generated numbers
average_numbers = sum(random_numbers)/ len(random_numbers)
average_numbers
```

```python
#e
# Calculate the average of the first 6 numbers in both sorted lists
average_ascending_first_6 = sum(sorted_numbers[:6]) / 6
average_descending_first_6 = sum(sorted_numbers_descending[:6]) / 6

average_ascending_first_6, average_descending_first_6
print(average_ascending_first_6, average_descending_first_6)

avg =  (average_ascending_first_6 + average_descending_first_6)/2
print('AVERAGE : ' , avg)
```

# Question 3 (20 points).

Create a Python program that encompasses various string manipulation task:

(a) Prompt the user to input two sentences.

(b) Transform each sentence into a list, where each element corresponds to a word in the sentence. For example, if the sentences are "Hello, how are you doing today?" and "This is an example sentence for the program!"

The program should output:
['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(c) Compare the two sentences. Print out a list of words that occur in both sentences.

(d) Remove any punctuation from the sentence and print the sanitized sentences. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”).

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
import string  # Import the string module for string constants like punctuation

def main():
    # (a) Prompt the user to input two sentences.
    sentence1 = input("Enter the first sentence: ")
    sentence2 = input("Enter the second sentence: ")

    # (b) Transform each sentence into a list of words and punctuation.
    sentence1_list = sentence1.split()  # Split the first sentence into a list of words.
    sentence2_list = sentence2.split()  # Split the second sentence into a list of words.
    combined_list = sentence1_list + sentence2_list  # Combine the lists for output.
    print(combined_list)

    # (c) Compare the two sentences and print out a list of words that occur in both.
    common_words = set(sentence1_list).intersection(set(sentence2_list))  # Use set intersection to find common words.
    print(list(common_words))  # Print the list of common words.

    # (d) Remove punctuation from the sentences and print the sanitized versions.
    translator = str.maketrans('', '', string.punctuation)  # Create a translation table for removing punctuation.
    sanitized_sentence1 = sentence1.translate(translator)  # Remove punctuation from the first sentence.
    sanitized_sentence2 = sentence2.translate(translator)  # Remove punctuation from the second sentence.
    print(sanitized_sentence1)
    print(sanitized_sentence2)

    # (e) Count the number of vowels in each of the sentences and print the counts.
    vowels = 'aeiou'  # Define the vowels.
    sentence1_vowels = {vowel: sanitized_sentence1.lower().count(vowel) for vowel in vowels}  # Count vowels in the first sentence.
    sentence2_vowels = {vowel: sanitized_sentence2.lower().count(vowel) for vowel in vowels}  # Count vowels in the second sentence.
    print(f"Sentence1 vowels: {sentence1_vowels}")
    print(f"Sentence2 vowels: {sentence2_vowels}")

if __name__ == "__main__":
    main()
```

# Question 4 (20 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
#a
def basic_calculator(expression):
    # Try to evaluate the basic mathematical expression using the eval function.
    # The eval function can evaluate simple python expressions.
    try:
        result = eval(expression)
        return result
    except Exception as e:
        # Return an error message if the expression is invalid.
        return f"Error: {str(e)}"

# Example usage
expression = "3 + 2 * 2 - 1 / 2"
result = basic_calculator(expression)
print(f"Result: {result}")
```

```python
def calculator_with_parentheses(expression):
    # This function is similar to the basic calculator but also correctly handles parentheses.
    # The eval function inherently supports the evaluation of expressions with parentheses.
    try:
        result = eval(expression)
        return result
    except Exception as e:
        return f"Error: {str(e)}"

# Example usage
expression = "(3 + 2) * (2 - 1) / 2"
result = calculator_with_parentheses(expression)
print(f"Result: {result}")
```

```python
import math

def enhanced_calculator(expression):
    # Define a dictionary of additional operations that can be used with eval.
    additional_operations = {
        'sqrt': math.sqrt,  # Square root
        'pow': pow,         # Exponentiation
        '%': lambda x, y: x % y  # Modulus operation
    }
    try:
        # Use eval to evaluate the expression, including the additional operations.
        result = eval(expression, {"__builtins__": None}, additional_operations)
        return result
    except Exception as e:
        return f"Error: {str(e)}"

# Example usage
expression = "sqrt(16) + pow(2, 3) % 3"
result = enhanced_calculator(expression)
print(f"Result: {result}")
```

```python
def calculator_with_error_handling(expression):
    # This function enhances the calculator by adding error handling for specific scenarios such as division by zero.
    try:
        # Evaluate the expression, including handling division by zero.
        result = eval(expression)
        return result
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed."
    except SyntaxError:
        return "Error: Invalid input expression."
    except Exception as e:
        return f"Error: {str(e)}"

# Example usage
expression = "1 / 0"  # Division by zero example
result = calculator_with_error_handling(expression)
print(f"Result: {result}")
```

```python
def run_calculator():
    while True:
        expression = input("Enter expression (or type 'exit' to quit): ")
        if expression.lower() == 'exit':
            print("Exiting calculator...")
            break
        result = calculator_with_error_handling(expression)
        print(f"Result: {result}")

# This function uses the calculator with error handling and includes a command to exit the program.
# To exit, the user types 'exit' when prompted for an expression.

#  the line below to run the calculator
run_calculator()
```

# Question 5 (30 points).

In the field of Data Science, data is often formatted as a comma-delimited (CSV) file, where each line in a file corresponds to a field's value. Refer to Chapter 9 (p. 106) for more information on comma-delimited files.

To answer this question, you will need to download the file [Assignment1_denton_housing.csv](https://raw.githubusercontent.com/unt-iialab/info5731-spring2022/main/assignments/Assignment1_denton_housing.csv) included with this assignment. The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (10 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (10 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
#a
# First, let's load the provided CSV file to understand its structure and contents.
import pandas as pd

# Load the CSV file into a DataFrame
file_path = 'Assignment1_denton_housing.csv'
housing_data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to understand its structure
housing_data.head()

```

```python
# Filter the data for "Occupied Housing Units" and then calculate the year-to-year difference.
occupied_housing = housing_data[housing_data['title_field'] == 'Occupied Housing Units'].sort_values(by='year')

# Calculate the difference from year to year
occupied_housing['difference'] = occupied_housing['value'].diff()

# Since the first year's difference will be NaN (as there's no previous year to compare), we'll drop that.
occupied_diff_cleaned = occupied_housing.dropna()

# Sort the differences in ascending order
occupied_diff_sorted = occupied_diff_cleaned.sort_values(by='difference')

# Format and print the differences
occupied_diff_sorted[['year', 'difference']]
```

```python
#b
# This involves calculating the percentages for vacant and occupied housing units for each year and formatting the output accordingly.

# Calculate the percentages for all years
# Group the data by 'year' to handle calculations for each year together
grouped_data = housing_data.pivot(index='year', columns='title_field', values='value')
grouped_data['Vacant Housing Units (%)'] = (grouped_data['Vacant Housing Units'] / grouped_data['Total Housing Units']) * 100
grouped_data['Occupied Housing Units (%)'] = (grouped_data['Occupied Housing Units'] / grouped_data['Total Housing Units']) * 100

# Select and rename the relevant columns for the final output
final_output = grouped_data[['Vacant Housing Units (%)', 'Occupied Housing Units (%)']]
final_output = final_output.rename(columns={'Vacant Housing Units (%)': 'Vacant Housing Units', 'Occupied Housing Units (%)': 'Occupied Housing Units'})

# Sort the DataFrame by year
final_output = final_output.sort_index(ascending=True)

# Reset the index to include 'year' as a column for display
final_output.reset_index(inplace=True)

final_output
```

```python
# Filter the dataset for rows where the title_field indicates 'Vacant Housing Units', sort these rows in descending order based on the 'value' column to get the year with the highest number of vacant housing units at the top, and select the top row using .iloc[0].
max_vacant_year = housing_data[housing_data['title_field'] == 'Vacant Housing Units'].sort_values(by='value', ascending=False).iloc[0]

# Similarly, filter the dataset for rows where the title_field is 'Occupied Housing Units', sort these rows in descending order by 'value' to get the year with the highest number of occupied housing units at the top, and select the top row using .iloc[0].
max_occupied_year = housing_data[housing_data['title_field'] == 'Occupied Housing Units'].sort_values(by='value', ascending=False).iloc[0]

# Print a heading for the output section concerning the highest number of vacant and occupied housing units.
print("\n(c) Year with the Highest Number of Vacant and Occupied Housing Units:")

# Print the year and value (number) of the highest number of vacant housing units, using string formatting to insert the year and value into the output string.
print(f"Vacant Housing Units: Year {max_vacant_year['year']}, Value {max_vacant_year['value']}")

# Print the year and value (number) of the highest number of occupied housing units, similarly using string formatting for the year and value.
print(f"Occupied Housing Units: Year {max_occupied_year['year']}, Value {max_occupied_year['value']}")
```

```python
from scipy.stats import hmean

# Filter the dataset to include only rows where the title_field is 'Total Housing Units'
total_housing_units = housing_data[housing_data['title_field'] == 'Total Housing Units']['value']

# Calculate the harmonic mean of the total housing units using the hmean function from scipy.stats
harmonic_mean_total_housing = hmean(total_housing_units)

harmonic_mean_total_housing
```
